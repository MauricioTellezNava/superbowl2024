```
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```
import os
```

```
df = pd.read_excel("drive/MyDrive/platicas/NFL.xlsx", header=1)
```

# Tampa Bay Bucaneers

```
data = df.dropna()
dataX =data[["Juego"]]
X_train = np.array(dataX)
y_train = data['Ofensiva'].values

regr = linear_model.LinearRegression()

regr.fit(X_train, y_train)

y_pred = regr.predict(X_train)

print('Coeficiente: \n', regr.coef_)
print('Bo: \n', regr.intercept_)
print("Error cuadraro medio: %.2f" % mean_squared_error(y_train, y_pred))
print('R2: %.2f' % r2_score(y_train, y_pred))
```

```
regr.predict([[20]])
```
